## Extracting the Dataset file from the zip file

```python
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() # extract the file to the home directory
zip_ref.close() # close the file
```

```python
import pandas as pd
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown']) # checking the file for null values and handling them
df['Ship Mode'].unique()
```

```python
df.columns
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns=df.columns.str.replace(' ','_')
```

```python
df.head(5) # renamed columns after formatting
```

```python
df['discount']=df['list_price']*df['discount_percent']*.01 # making a discount column
```

```python
df.head(5)
```

```python
df['sale_price']=df['list_price']-df['discount'] # sale price column
df['profit']=df['sale_price']-df['cost_price'] # profit column
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d") # formatting the dates from object data type to datetime
```

```python
df.head(5)
```

```python
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df # applying the changes to the original dataset
```
